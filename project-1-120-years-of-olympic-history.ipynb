# Exploring dataset

```python
df_athlete.head()
```

```python
df_athlete.info()
```

```python
# Number of null values
df_athlete.isnull().sum()
```

```python
# Droping null values
df_athlete = df_athlete.dropna()
```

```python
# group by teams, counting medals and sorting values 
# Convering to data frame and selecting top 10 values
dist_medals = df_athlete.groupby('Team')['Medal'].count().sort_values(ascending = False).reset_index().head(10)
dist_medals
```

# 1. Plot the distribution of medals by countries


```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize = (20,5))
sns.barplot(x='Team',y = 'Medal',data=dist_medals)
plt.title('Distribution of medals by country')
plt.xlabel('Country')
plt.ylabel('Number of medals')
plt.show()
```

# 2. Plot the number of athletes in each sports

```python
sport_athlete = df_athlete.groupby('Sport')['Name'].nunique().sort_values(ascending=False).reset_index(name = 'No of athletes').head(10)
```

```python
plt.figure(figsize = (20,5))
sns.barplot(x='Sport',y='No of athletes',data = sport_athlete)
plt.title('Number of athletes in each sports')
plt.xlabel('Sport')
plt.ylabel('Number of athletes')
plt.show()
```

# 3. Top countries with most gold medals


```python
df_gold = df_athlete[df_athlete['Medal']=='Gold']
df_gold
```

```python
df_gold = df_gold.Team.value_counts().reset_index(name = 'gold medals').head(10)
df_gold
```

# 4. Plot the top countries with most gold medals


```python
plt.figure(figsize = (20,5))
sns.barplot(x='Team',y='gold medals',data = df_gold)
plt.title('Number of gold medals by countries')
plt.xlabel('Team')
plt.ylabel('Number of gold medals')
plt.show()
```

# 5. Convert the "year" column to "datetime"


```python
df_athlete.info()
```

```python
df_athlete['Year'] = pd.to_datetime(df_athlete['Year'],format = "%Y").dt.year
```

```python
df_athlete
```

# 6. Plot the number of medals over time

```python
year = df_athlete.Year.value_counts().reset_index()
year = year.sort_values(by='Year')
year.head()
```

```python
plt.figure(figsize = (20,5))
sns.lineplot(x='Year',y = 'count', data = year)
plt.title('Number of medals ')
plt.xlabel('Year')
plt.ylabel('Number of medals')
plt.show()
```
